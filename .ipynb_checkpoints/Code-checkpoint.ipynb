```python
# Drop Columns 

drop_col = [
    'ID','Name','SSN'
]

data = data.drop(columns=drop_col, errors='ignore')
```

```python
# Change to numerical
cols = ['Age', 'Num_of_Delayed_Payment', 'Num_of_Loan']

for col in cols:
    data[col] = data[col].astype(str).str.extract('(\d+)').astype(float)

    
```

```python
# Remove Garbage

data = data[data['Payment_Behaviour']!='!@9#%8']
data = data[data['Occupation']!='_______']
data = data[data['Credit_Mix']!='_']
data = data[data['Delay_from_due_date']>=0]
```

```python
# Make the History age into Month
years = data['Credit_History_Age'].astype(str).str.split(' ').str[0]
months = data['Credit_History_Age'].astype(str).str.split(' ').str[3]

years = pd.to_numeric(years, errors='coerce')
months = pd.to_numeric(months, errors='coerce')

data['Credit_History_Age_Month'] = years * 12 + months
```

```python
# Remove garbage from columns and change thier type into numric

num_cols = ['Num_Bank_Accounts','Num_Credit_Card','Outstanding_Debt','Annual_Income','Num_of_Delayed_Payment','Total_EMI_per_month','Amount_invested_monthly','Interest_Rate', 'Num_Credit_Inquiries']

for col in num_cols:
    data[col] = data[col].astype(str)
    data[col] = data[col].str.replace('_','', regex=False)
    data[col] = data[col].str.replace(' ','', regex=False)
    data[col] = data[col].str.replace('[^0-9\.]','', regex=True)
    data[col] = data[col].replace('', np.nan)
    data[col] = pd.to_numeric(data[col], errors='coerce')
```

```python
# Fille the missing values with the median 
for col in num_cols:
    median = data[col].median()
    data[col] = data[col].fillna(median)
```

### Data Handling for Modeling
- __Transform, encode, and prepare features__
    - Categorical encoding
    - Feature scaling (especially for KNN)


```python
# Feature Engineering

data['Total_Num_Accounts'] = data['Num_Bank_Accounts'] + data['Num_Credit_Card']
data['Debt_Per_Account'] = data['Outstanding_Debt'] / data['Total_Num_Accounts']
data['Debt_to_Income_Ratio'] = data['Outstanding_Debt'] / data['Annual_Income']
data['Delayed_Payments_Per_Account'] = data['Num_of_Delayed_Payment'] / data['Total_Num_Accounts']
data['Total_Monthly_Expenses'] = data['Total_EMI_per_month'] + data['Amount_invested_monthly']
```

```python
# Encoding

credit_map = {'Bad':0, 'Standard':1, 'Good':2}
payment_map = {
    'Low_spent_Small_value_payments':1,
    'Low_spent_Medium_value_payments':2,
    'Low_spent_Large_value_payments':3,
    'High_spent_Small_value_payments':4,
    'High_spent_Medium_value_payments':5,
    'High_spent_Large_value_payments':6
}
data['Credit_Mix_Enc'] = data['Credit_Mix'].map(credit_map)
data['Payment_Behaviour_Enc'] = data['Payment_Behaviour'].map(payment_map)
data['Credit_Score_Enc'] = data['Credit_Score'].map({'Poor':0,'Standard':1,'Good':2})
```

```python
# Combining Features

data['Total_Num_Accounts'] = data['Num_Bank_Accounts'] + data['Num_Credit_Card']

data['Debt_Per_Account'] = np.where(
    data['Total_Num_Accounts'] > 0,
    data['Outstanding_Debt'] / data['Total_Num_Accounts'],
    0
)

data['Debt_to_Income_Ratio'] = np.where(
    data['Annual_Income'] > 0,
    data['Outstanding_Debt'] / data['Annual_Income'],
    0
)

data['Delayed_Payments_Per_Account'] = np.where(
    data['Total_Num_Accounts'] > 0,
    data['Num_of_Delayed_Payment'] / data['Total_Num_Accounts'],
    0
)

data['Total_Monthly_Expenses'] = data['Total_EMI_per_month'] + data['Amount_invested_monthly']
```

```python
# X,y, Scaling and SMOTE
X = data.drop(columns=['Credit_Score'])
X = pd.get_dummies(X, drop_first=True)
y = data['Credit_Score']

sc = StandardScaler()
sm = SMOTE(random_state=42)
```

## KNN Version 1:

```python
# Train_Test_Spli
X_train3, X_test3, y_train3, y_test3 = train_test_split(X1, y, test_size=0.25, random_state=42, stratify=y)
```

```python
# Scaling
X_train_sc3 = sc.fit_transform(X_train3)
X_test_sc3  = sc.transform(X_test3)
X_train_sc3, y_train3 = sm.fit_resample(X_train_sc3, y_train3)
```

## KNN Version 2:

```python
# X, y

X3 = data[['Age','Outstanding_Debt','Num_of_Delayed_Payment','Credit_Utilization_Ratio',
           'Annual_Income','Payment_Behaviour','Payment_of_Min_Amount','Credit_History_Age']]
y = data['Credit_Score_Enc']


X3 = pd.get_dummies(X3, drop_first=True, dtype='float32')

X3 = X3.replace([np.inf, -np.inf], np.nan)
X3 = X3.fillna(0.0)
```

```python
# Train_Test_Split

X_train4, X_test4, y_train4, y_test4 = train_test_split(X3, y, test_size=0.25, random_state=42, stratify=y)
```

```python
# Pipeline
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('knn', KNeighborsClassifier())
])
```

```python
# GricSearchCV
pgrid = {
    'knn__n_neighbors': np.arange(1, 31, 5)
}
```

```python
# GricSearchCV
gs = GridSearchCV(
    estimator=pipe,
    param_grid=pgrid,
    cv=10,
    scoring='accuracy',
    n_jobs=8,  
    refit=True
)
```

```python
# Train The Model
gs.fit(X_train4, y_train4)
```

```python
# Best K number and CV

best_params = gs.best_params_
best_cv_r2  = gs.best_score_

print("Best params from GridSearchCV:", best_params)
print("Best CV R2 from GridSearchCV:", best_cv_r2)
```

```python
# Predction

best = gs.best_estimator_

y_pred4 = best.predict(X_test4)
```

### Model Evaluation
__Use appropriate metrics based on task type and compare between the Models and their versions__

__For Regression:__
- R² Score
- Root Mean Squared Error (RMSE)
- Residual plots


__For Classification:__
- Accuracy
- Precision
- Recall
- F1 Score
- ROC-AUC _**(Optional)**_
- Confusion Matrix

_Include visualizations and interpretation for each metric._


```python
# Models Evaluation
results = []
models = {
    'Logistic Regression V1': (y_test1, y_pred1),
    'Logistic Regression V2': (y_test2, y_pred2),
    'KNN Classifier V1': (y_test3, y_pred3),
    'KNN Classifier V2': (y_test4, y_pred4)
}

for name, (yt, yp) in models.items():
    acc = accuracy_score(yt, yp)
    prec = precision_score(yt, yp, average='macro', zero_division=0)
    rec = recall_score(yt, yp, average='macro', zero_division=0)
    f1 = f1_score(yt, yp, average='macro', zero_division=0)
    results.append([name, acc, prec, rec, f1])

    print(f"\n=== {name} ===")
    print(f"Accuracy: {acc:.3f}")
    print(f"Precision: {prec:.3f}")
    print(f"Recall: {rec:.3f}")
    print(f"F1-Score: {f1:.3f}")
    print("-"*30)
    cm = confusion_matrix(yt, yp)
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='g', cmap='Blues')
    plt.title(name + " — Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

eval_df = pd.DataFrame(results, columns=['Model Version','Accuracy','Precision','Recall','F1-Score'])
display(eval_df)
```

```python
# Plot the Evaluations
eval_long = eval_df.melt(id_vars='Model Version', var_name='Metric', value_name='Score')
plt.figure(figsize=(12,6))
sns.barplot(data=eval_long, x='Metric', y='Score', hue='Model Version', palette='coolwarm')
plt.ylim(0,1)
plt.title('Model Performance Comparison')
plt.ylabel('Score (0–1)')
plt.legend(title='Model Version')
plt.show()
```
